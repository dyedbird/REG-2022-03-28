```python
import requests, bs4
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
url = 'https://fbref.com/en/comps/26/Super-Lig-Stats#results113151_overall'
response = requests.get(url)
status = response.status_code
if status == 200:
  page = response.text
  soup = bs(page)
else:
  print(f"Oops! Received status code {status}")
```

```python
league_table_22 = []

ranking= [rank['csk'] for rank in soup.find_all\
                            ('th', attrs={'scope':'row','data-stat':'rank'})][0:21]

squad = [team.text.strip() for team in soup.find_all('td',attrs={'data-stat':'squad'})][0:21]

wins = [win.text for win in soup.find_all('td',attrs = {'data-stat':'wins'})][0:21]

draws = [draw.text for draw in soup.find_all('td',attrs = {'data-stat':'draws'})][0:21]

losses = [loss.text for loss in soup.find_all('td',attrs = {'data-stat':'losses'})][0:21]

goals_for = [goal.text for goal in soup.find_all('td',\
                                                    attrs = {'data-stat':'goals_for'})][0:21]

goals_against = [goal.text for goal in soup.find_all\
                                         ('td',attrs = {'data-stat':'goals_against'})][0:21]

points = [point.text for point in soup.find_all('td',attrs = {'data-stat':'points'})][0:21]

x_goals = [goal.text for goal in soup.find_all('td',attrs = {'data-stat':'xg_for'})][0:21]

x_goals_A = [goal.text for goal in soup.find_all\
                                             ('td',attrs = {'data-stat':'xg_against'})][0:21]

attendance = [att.text for att in soup.find_all\
                                    ('td',attrs = {'data-stat':'attendance_per_g'})][0:21]


for num in range(0,20):
    my_dict={'RANK':ranking[num],'TEAM':squad[num], 'W':wins[num], 'D':draws[num],\
             'L':losses[num], 'GF':goals_for[num], 'GA':goals_against[num],\
             'PTS':points[num]}
    league_table_22.append(my_dict)
    my_dict = dict()

league_table_22
```

```python
SL_22_table_df = pd.DataFrame(league_table_22)
SL_22_table_df.style.hide_index()
```
